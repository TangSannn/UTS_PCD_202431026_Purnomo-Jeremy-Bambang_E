```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 202431026_Purnomo Jeremy Bambang
# 2. BAGIAN A
# Membaca Gambar Asli

gambar_bgr = cv2.imread('Foto_backlight.jpeg')

# Konversi BGR ke RGB untuk ditampilkan dengan matplotlib
gambar_rgb = cv2.cvtColor(gambar_bgr, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(9, 5))
plt.imshow(gambar_rgb)
plt.title('Gambar Asli')
plt.axis('off')
plt.show()

# Menyimpan gambar asli
cv2.imwrite('2. A. Gambar_Asli.jpeg', gambar_bgr)
```

```python
# 202431026_Purnomo Jeremy Bambang
# 2. BAGIAN B
# Konversi Gambar Menjadi Grayscale

gambar_gray = cv2.cvtColor(gambar_bgr, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(9, 5))
plt.imshow(gambar_gray, cmap='gray')
plt.title('Gambar Gray')
plt.axis('off')
plt.show()

# Menyimpan gambar gray
cv2.imwrite('2. B. Gambar_Gray.jpeg', gambar_gray)
```

```python
# 202431026_Purnomo Jeremy Bambang
# 2. BAGIAN C
# Meningkatkan Kecerahan Gambar Gray

# Menambah nilai piksel agar gambar lebih terang
brightness = 55
gambar_gray_dicerahkan = cv2.add(gambar_gray, brightness)

plt.figure(figsize=(9, 5))
plt.imshow(gambar_gray_dicerahkan, cmap='gray')
plt.title('Gambar Gray yang Dipercerah')
plt.axis('off')
plt.show()

# Menyimpan gambar gray yang dicerahkan
cv2.imwrite('2. C. Gambar_Gray_Dicerahkan.jpeg', gambar_gray_dicerahkan)
```

```python
# 202431026_Purnomo Jeremy Bambang
# 2. BAGIAN D
# Meningkatkan Kontras Gambar Gray

# Membuat copy gambar gray
gambar_gray_kontras = gambar_gray.copy().astype(np.float32)

# Nilai kontras
alpha = 1.6

# Rumus kontras manual:
# nilai baru = nilai lama * alpha
gambar_gray_kontras = gambar_gray_kontras * alpha

# Membatasi nilai agar tetap 0 sampai 255
gambar_gray_kontras = np.clip(gambar_gray_kontras, 0, 255)

# Mengubah kembali ke uint8
gambar_gray_kontras = gambar_gray_kontras.astype(np.uint8)

# Menampilkan hasil
plt.figure(figsize=(9, 5))
plt.imshow(gambar_gray_kontras, cmap='gray')
plt.title('Gambar Gray yang Diperkontras')
plt.axis('off')
plt.show()

# Menyimpan gambar gray yang diperkontras
cv2.imwrite('2. D. Gambar_Gray_Diperkontras.jpeg', gambar_gray_kontras)
```

```python
# 202431026_Purnomo Jeremy Bambang
# 2. BAGIAN E
# Meningkatkan Kecerahan dan Kontras Gambar Gray

gambar_gray_cerah_kontras = gambar_gray.copy().astype(np.float32)

# Area gelap, sedang, dan terang
area_gelap = gambar_gray < 100
area_sedang = (gambar_gray >= 100) & (gambar_gray < 170)
area_terang = gambar_gray >= 170

# Area gelap dibuat lebih terang dan lebih kontras
gambar_gray_cerah_kontras[area_gelap] = gambar_gray_cerah_kontras[area_gelap] * 1.7 + 75

# Area sedang dinaikkan secukupnya
gambar_gray_cerah_kontras[area_sedang] = gambar_gray_cerah_kontras[area_sedang] * 1.2 + 30

# Area terang dikurangi sedikit agar tidak terlalu terbakar
gambar_gray_cerah_kontras[area_terang] = gambar_gray_cerah_kontras[area_terang] * 0.85

# Membatasi nilai piksel agar tetap 0 sampai 255
gambar_gray_cerah_kontras = np.clip(gambar_gray_cerah_kontras, 0, 255)

# Mengubah kembali ke uint8
gambar_gray_cerah_kontras = gambar_gray_cerah_kontras.astype(np.uint8)

# Menampilkan hasil
plt.figure(figsize=(9, 5))
plt.imshow(gambar_gray_cerah_kontras, cmap='gray')
plt.title('Gambar Gray yang Dipercerah dan Diperkontras')
plt.axis('off')
plt.show()

# Menyimpan gambar gray yang dicerahkan dan diperkontras
cv2.imwrite('2. E. Gambar_Gray_Dicerahkan_Diperkontras.jpeg', gambar_gray_cerah_kontras)
```
